```python
# CID CAPITULOS

import pandas as pd
import numpy as np


df = pd.read_csv("grupos.csv")
# header=None

# Change all data types to strings
df = df.astype(str)

def remove_line_breaks(string):
    """
    Remove line breaks from a string.

    Args:
        string (str): The input string.

    Returns:
        str: The cleaned string.
    """
    string = string.replace('\r\n', ' ').replace('\n', ' ')
    return string

df = df.applymap(remove_line_breaks)
df = df.replace('"', '', regex=True)

df[['codigo', 'descricao', 'descricao_breve']] = df['codigo;"descricao";"descricao_breve"'].str.split(';', expand=True)
df = df[['codigo', 'descricao', 'descricao_breve']]

df.to_csv('out.csv', index=False) 
```

```python
def clean_string(string):
    """
    Remove leading and trailing whitespaces and quotes from a string.

    Args:
        string (str): The input string.

    Returns:
        str: The cleaned string.
    """
    # Remove leading and trailing whitespaces
    string = string.strip()

    # Define the characters to remove
    chars_to_remove = ["'", '"']

    # Remove any remaining quotes
    for char in chars_to_remove:
        while string.startswith(char) or string.endswith(char):
            string = string.strip(char)

    return string
```
